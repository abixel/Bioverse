```python
# Import statements
import emcee
import sys
sys.path.append('../')
from bio import *
%matplotlib inline

# Pyplot parameters
plt.rcParams['font.size'] = 20.
figsize = (16,6)
```

```python
# Simulate a dataset for a LUVOIR-like survey with f_water_habitable = 0.8 and f_water_nonhabitable = 0.05
generator = classes.Generator('LUVOIR_EPOS')
survey = classes.Survey('LUVOIR')
sample = generator.generate(f_water_habitable=0.5,f_water_nonhabitable=0.05)
detected = survey.compute_yield(sample,simple=True)
data = survey.observe_simple(detected)

#data = data[:int(0.25*len(data))]

# Compute insolation from semi-major axis
data['S'] = data['L_st']/data['a']**2
```

```python
# Plot (for visualization) the fraction of planets with water vapor vs. insolation
bins, values, errors = analysis.binned_average(np.log10(data['S']),data['has_H2O'],bins=20,match_counts=True)
x, dx = (bins[1:]+bins[:-1])/2., (bins[1:]-bins[:-1])/2.
y, dy = values, errors

fig, ax = plt.subplots(figsize=figsize)
ax.errorbar(x,y,xerr=dx,yerr=dy,linestyle='None',marker='s')
ax.set_xlabel('log$_{10}$(S)',fontsize=28)
ax.set_ylabel('$f_{H_2O}$',fontsize=28)
ax.axvline(np.log10(1.15),c='black',lw=2)
ax.axvline(np.log10(0.37),c='black',lw=2)
plt.show()
```

```python
# Compute the Mann-Whitney p-value comparing the abundance of H2O inside vs outside of the habitable zone
p = analysis.detect_hz_water(data,S_inner=1.15,S_outer=0.37)
print("Likelihood of the null hypothesis (i.e. no habitable zone): {:.1E}".format(p))
```

```python
# Example 1: Vary S_inner to determine which value minimizes p (fixed S_outer)
S_inner_grid = 10**np.arange(-1,1,0.01)
p_inner,U_inner = np.ones((2,len(S_inner_grid)))
for i,S_inner in enumerate(S_inner_grid):
    p_inner[i],U_inner[i] = analysis.detect_hz_water(data,S_inner=S_inner,S_outer=0.37,return_U=True)

# Example 2: Vary S_outer to determine which value minimizes p (fixed S_inner)
S_outer_grid = 10**np.arange(-2,0,0.01)
p_outer,U_outer = np.ones((2,len(S_outer_grid)))
for i,S_outer in enumerate(S_outer_grid):
    p_outer[i],U_outer[i] = analysis.detect_hz_water(data,S_inner=1.15,S_outer=S_outer,return_U=True)

# Example 3: Vary S_inner and S_outer together to minimize p
p_both,U_both = np.ones((2,len(S_inner_grid),len(S_outer_grid)))
for i,S_inner in enumerate(S_inner_grid):
    for j,S_outer in enumerate(S_outer_grid):
        p_both[i,j],U_both[i,j] = analysis.detect_hz_water(data,S_inner=S_inner,S_outer=S_outer,return_U=True)

# Print the best-fit values from the third example
idx = np.unravel_index(p_both.argmin(),p_both.shape)
S_inner_best,S_outer_best = S_inner_grid[idx[0]],S_outer_grid[idx[1]]

print("Best-fit values:")
print("S_inner = {:.2f}".format(S_inner_best))
print("S_outer = {:.2f}".format(S_outer_best))
```

```python
# Plot the results for the three examples
fig,ax = plt.subplots(1,3,figsize=figsize)

ax[0].plot(S_inner_grid,p_inner)
ax[0].axvline(1.15,c='black',lw=2)
ax[0].axvline(S_inner_grid[np.argmin(p_inner)],lw=2,c='red')
ax[0].set_xscale('log')
ax[0].set_yscale('log')
ax[0].set_xlabel('S_inner')
ax[0].set_ylabel('p value')
ax[0].set_title('Inner edge only')

ax[1].plot(S_outer_grid,p_outer)
ax[1].axvline(0.37,c='black',lw=2)
ax[1].axvline(S_outer_grid[np.argmin(p_outer)],lw=2,c='red')
ax[1].set_xscale('log')
ax[1].set_yscale('log')
ax[1].set_xlabel('S_outer')
ax[1].set_title('Outer edge only')

extent = np.log10([min(S_outer_grid),max(S_outer_grid),min(S_inner_grid),max(S_inner_grid)])
extent[[1,3]] += 0.1
im = ax[2].imshow(np.log10(p_both),extent=extent)
idx = np.unravel_index(p_both.argmin(),p_both.shape)
ax[2].scatter(np.log10(S_outer_grid[idx[1]]),np.log10(S_inner_grid[idx[0]]),c='red',s=50)
plt.colorbar(im)
ax[2].set_xlabel('log(S_outer)')
ax[2].set_ylabel('log(S_inner)')
ax[2].set_title('Both')

plt.subplots_adjust(top=0.8,wspace=0.4)
plt.savefig('measurement.pdf')
plt.suptitle('Measured values (red) vs true values (black)')
```
